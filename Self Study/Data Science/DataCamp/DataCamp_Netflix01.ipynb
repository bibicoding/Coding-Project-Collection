```python
# Importing pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
# preview the data info
print(netflix_df)
```

```python
# Subset a DataFrame for movies released in the 1990s
netflix_df_oldies = netflix_df[np.logical_and(np.logical_and(netflix_df["release_year"]>=1990, netflix_df["release_year"]<=1999), netflix_df["type"]=="Movie")]

# Print out the movies concerned
print(netflix_df_oldies.loc[:, ["title", "type", "release_year"]])
```

```python
# Visualize the most frequent movie duration
plt.hist(netflix_df_oldies["duration"], bins=100)
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Durations")
plt.xticks(np.arange(0, 200, 10))
plt.grid()
plt.show()

# Approximate most frequent movie duration
duration = netflix_df_oldies["duration"].value_counts().idxmax()
print(duration)
```

```python
# Count the number of short action movies
netflix_df_shortaction = netflix_df_oldies[np.logical_and(netflix_df_oldies["genre"]=="Action", netflix_df_oldies["duration"] < 90)]

# Print out the movies concerned
print(netflix_df_shortaction.loc[:, ["title"]])

# Count the movies in the DataFrame
short_movie_count = len(netflix_df_shortaction)

# Print out the number of short action movies
print(short_movie_count)
```
